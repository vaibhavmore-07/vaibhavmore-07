```python
df = pd.read_csv( '/content/drive/MyDrive/achive/Cuisine_rating (1).csv')
```

```python
df.head()
```

```python
def set_name(x):
    return x.rstrip().replace(' ','_').lower()

df.rename(mapper = set_name, axis = 1,inplace = True)
df.columns
```

```python
df.drop_duplicates(subset = 'user_id', keep = False, inplace = True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['city'] = np.array(map(lambda x: x.split(',')[-1], df['location']))
df.city.value_counts()
```

```python
df.columns
```

```python
le = LabelEncoder()
df['le_alcohol'] = le.fit_transform(df.alcohol)
df['le_smoker'] = le.fit_transform(df.smoker)
df['le_often'] = le.fit_transform(df.often_a_s)

df.head()
```

```python
df.food_rating.value_counts()
```

```python
food_service_dict = {}
for food in df['food_rating']:
    service_count = {}

    food_tmp = df.loc[df['food_rating'] == food]

    for service in df['service_rating']:
        service_tmp = food_tmp.loc[food_tmp['service_rating'] == service]

        service_count[service] = service_tmp.shape[0] / food_tmp.shape[0]

    food_service_dict[food] = {key: val for key, val in sorted(service_count.items(), key = lambda x: x[0])}

food_service_dict = {key: val for key, val in sorted(food_service_dict.items(), key = lambda x: x[0])}
```

```python
food_service_dict
```

```python
food_service = pd.DataFrame.from_dict(food_service_dict)

food_service.index.name = 'Service Rating'
food_service.columns.name = 'Food Rating'

food_service = food_service.transpose()
food_service
```

```python
plt.style.use('ggplot')
food_service.plot(
                  kind = 'bar',
                  stacked = True,
                  title = 'Percentage of Service Rating grouped by Food Rating',
                  mark_right = True)

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.8), title = 'Service Rating')
plt.xticks(rotation = 'horizontal')
plt.show()
```

```python
activity_budget_dict = {}

for ac in df.activity.unique():
    activity_ = {}
    activity_tmp = df.loc[df['activity'] == ac]

    for bud in df.budget.unique():
        budget_tmp = activity_tmp.loc[activity_tmp['budget'] == bud]
        activity_[bud] = budget_tmp.shape[0]/activity_tmp.shape[0]

    activity_ = {key: val for key, val in sorted(activity_.items(), key = lambda x: x[0])}
    activity_budget_dict[ac] = activity_

activity_budget_dict
```

```python
activity_budget = pd.DataFrame(activity_budget_dict)
activity_budget = activity_budget.transpose()
activity_budget['Total'] = sum(activity_budget[x] for x in range(1, 6))
activity_budget
```

```python
plt.style.use('ggplot')

activity_budget.plot(y = range(1, 6),
                     kind = 'bar',
                     stacked = True)
plt.title('Activity: Budget Distribution')
plt.xticks(rotation = -45)
plt.xlabel('Activity')
plt.ylabel('Budget')
plt.show()
```
